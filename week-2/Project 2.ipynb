```python
print("This is a program that can solve quadratic, cubic and quartic equations, based on their coefficients.")
print("What type of equation are you solving?")
print("1. ax²+bx+c=0\n2. ax³+bx²+cx+d=0\n3. ax⁴+bx³+cx²+dx+e=0")
mode=0

def inputInt():
    try:
        return int(input())
    except:
        return inputInt()


def inputIntList(items):
    list=[]
    while items>0:
        list.append(inputInt())
        items-=1
    return list
    

while True:
    mode=inputInt()
    if 0<=mode<=3 :
        break

if mode==1:
    print("Enter coefficients of the equation in the order a,b,c.")
    [a,b,c]=inputIntList(3)
    x1=-b+(b**2-4*a*c)**.5/(2*a)
    x2=-b-(b**2-4*a*c)**.5/(2*a)
    print(f"The roots are {x1} and {x2}")


if mode==2:
    print("Enter coefficients of the equation in the order a,b,c,d.")
    [a,b,c,d]=inputIntList(4)
    # Based on Cardo's formula
    x=-b/(3*a)+((9*a*b*c-2*b**3-27*a**2*d)/(54*a**3)+(((9*a*b*c-2*b**3-27*a**2*d)/(54*a**3))**2+((3*a*c-b**2)/(9*a**2))**3)**0.5)**(1/3)+((9*a*b*c-2*b**3-27*a**2*d)/(54*a**3)-(((9*a*b*c-2*b**3-27*a**2*d)/(54*a**3))**2+((3*a*c-b**2)/(9*a**2))**3)**0.5)**(1/3)
    x2=(-0.5+(3**(1/3))*1j/2)*x
    x3=(-0.5-(3**(1/3))*1j/2)*x
    print(f"The roots are {x} , {x2} , {x3}.")

if mode==3:
    print("Enter coefficients of the equation in the order a,b,c,d,e.")
    [a,b,c,d,e]=inputIntList(5)
    # This is based on Ferrari's method.
    p=(8*a*c-3*b**2)/(8*a**2)
    q=(b**3-4*a*b*c+8*a**2*d)/(8*a**3)
    r0 = c**2-3*b*d+12*a*e
    r1 = 2*c**3 - 9*b*c*d + 27*b**2*e + 27*a*d**2 - 72*a*c*e
    Q = ((r1+(r1**2 - 4*r0**3)**0.5)/2)**(1/3)
    S = (-2*p/3+(Q+r0/Q)/(3*a))**0.5/2
    x1=-b/(4*a)-S+((-4*S**2-2*p+q/S)**0.5)/2
    x2=-b/(4*a) - S - ((-4*S**2 - 2*p + q/S)**0.5)/2
    x3=-b/(4*a) + S + ((-4*S**2 - 2*p - q/S)**0.5)/2
    x4=-b/(4*a) + S - ((-4*S**2 - 2*p - q/S)**0.5)/2
    print(f"The roots are {x1} , {x2} , {x3}, and {x4}.")
    
```
