```python
def inputInt():
    try:
        return int(input())
    except:
        return inputInt()


print("Hello. This is a program for Ifzin Technology.\nThis program is being run to determine your ANNUAL TAX REVENUE.\nEnter your age and your years of experience.")
age=inputInt()
yoe=inputInt()
atr=0

if age>=55 and yoe>25 :
    atr=5600000
elif age>=45 and yoe>20 :
    atr=4480000
elif age>=35 and yoe>10 :
    atr=1500000
elif age<35 and yoe<10 :
    atr=550000

print(f"Your Annual Tax Revenue is {atr}.")
```

#### BEGIN
- INPUT the user's age and years of experience, as age and youe
- SET atr(Annual tax revenue) to 0.
- IF age>=55 and yoe>25 SET atr=5600000
- ELSE IF age>=45 and yoe>20 SET atr=4480000
- ELSE IF age>=35 and yoe>10 SET atr=1500000
- ELSE IF age<35 and yoe<10 SET atr=550000
- PRINT atr (Annual tax revenue)
#### END
