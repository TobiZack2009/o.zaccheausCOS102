```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("titanic.csv")
count=0
print(data.isnull().sum().sum()/len(data))
```

Since about 97% of the data is is missing, we need to fill in the data.
For age, we can fill it with median, since that is not affected from outliers.
Though the cabin column has blanks, there is no way we can fill it mathematically. So we fill it with a blank string.

```python
data.fillna({"Age": data["Age"].median() },inplace=True)
data.fillna({"Cabin": "" },inplace=True)
print("Filled in missing data!")
data

```

We can now plot, now that we have filled in missing data. 
Let's start with a basic line chart.

```python
data.plot(kind="line")
plt.show()
```

We use a line graph because it shows us the relationship between the items best. Let's do some more plots. 

We can look for outliers on price, for example. Let's use the box plot on price.

```python
data["Age"].plot(kind="box",
                 title="Price Box Plot",
                 ylabel="Price (USD)",
                 patch_artist=True,
                 showfliers=False,
                 whiskerprops={'color': 'darkblue'},
                 capprops={'color': 'blue'},
                 boxprops={'facecolor': 'lightgreen', 'edgecolor': 'darkgreen'})
plt.show()
```

Here, we can see there are outliers.
