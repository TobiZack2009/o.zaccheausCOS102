```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("AB_NYC_2019.csv")
count=0
print(data.isnull().sum().sum()/len(data))
```

Since about 40% of the data is is missing, we need to fill in the data, including dates.
We can fill minimum nights and reviews per month using median, as the data does not have outliers.
We fill the dates and names using 0. There is no way to mathematically interpolate names, and for dates we fill using 0.

```python
data.fillna({"minimum_nights": data["minimum_nights"].median() },inplace=True)
data.fillna({"reviews_per_month": data["reviews_per_month"].median() },inplace=True)
data.fillna({"name": "" },inplace=True)
data.fillna({"late_review": 0 },inplace=True)
print("Filled in missing data!")
data


```

We can now plot, now that we have filled in missing data.

```python
data.plot(kind="line")
plt.show()
```

We use a line graph because it shows us the relationship between the items best. Let's do some more plots. 

We can look for outliers on price, for example. Let's use the box plot on price.

```python
data["price"].plot(kind="box",
                 title="Price Box Plot",
                 ylabel="Price (USD)",
                 patch_artist=True,
                 showfliers=False,
                 whiskerprops={'color': 'darkblue'},
                 capprops={'color': 'blue'},
                 boxprops={'facecolor': 'lightgreen', 'edgecolor': 'darkgreen'})
plt.show()
```

Here, we can see there are outliers in price, at close to 0 and close to 350.
