```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("AB_NYC_2019.csv")
count=0
print(data.isnull().sum().sum()/len(data))
```

Since about 40% of the data is is missing, we need to fill in the data.
We can fill the data using median, as the data does not have outliers.

```python
data.fillna({"minimum_nights": data["minimum_nights"].median() })
print("Filled in missing data!")
data


```

We can now plot, now that we have filled in missing data.

```python
data.plot(kind="line")
plt.show()
```
